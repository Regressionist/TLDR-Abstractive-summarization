```python
import numpy as np
import spacy
import os
from collections import Counter
import torch
import glob
from spacy import attrs
```

```python
nlp = spacy.load('en') # loads default English object
cnn_dir = 'data/cnn/'
cnn_tok_dir = 'data/cnn_tokenized/'
```

```python
file_list = [os.path.join(cnn_dir,file) for file in os.listdir(cnn_dir)]
```

```python
counter=1
for file in file_list:
    with open(file) as f:
        text=f.read()
    a=[x.text for x in list(nlp(text))]
    new_text=' '.join(a)
    
    with open(cnn_tok_dir+'{}.story'.format(counter),'w') as f:
        f.write(new_text)
    counter=counter+1
    if (counter%100)==0:
        print ('Step: {}'.format(counter))
```
